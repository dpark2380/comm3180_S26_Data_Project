# Quality of Life Per Dollar
## Data Journalism Final Project

## Research Question

Which cities offer the highest quality of urban life relative to what it costs to live there, and what do they have in common?

Rather than reproducing existing liveability rankings (which are built for wealthy expats), we build an original "value for living" index that centres affordability alongside quality.

## Research Sub-Questions

1. Which cities score highest on quality of life relative to cost of living?
2. Are sweet spot cities clustered in particular world regions, and does national wealth explain the pattern?
3. Which dimension (health, transport, or housing) most strongly drives overall quality of life?
4. Does income inequality explain why some low-cost cities still score poorly on quality of life?
5. Has the cost of living risen faster than quality of life since COVID?

## Analysis Plan

### Section 1 : Introduction and Framing

- Show that the top 10 most liveable cities (EIU 2025) are also among the world's most expensive
- Introduce the "value for living" concept as a more democratic measure
- Overview of cities and regions covered in the analysis

### Section 2 : Building the Index

- Combine Numbeo sub-indices (Safety, Healthcare, Pollution, Climate, Traffic) into a composite quality score, keeping cost of living completely separate
- Divide composite quality score by Cost of Living Index to produce a "value for living" ratio per city
- Bar chart of top 30 cities by value index
- Heatmap showing how each city scores across individual dimensions

### Section 3 : Geographic Patterns

- Choropleth world map coloured by value for living index
- Regional breakdown: which parts of the world produce the most sweet spot cities?
- Scatter plot of value index vs. GDP per capita (World Bank): looking for the middle-income sweet spot

### Section 4 : Dimension Deep-Dive

- Compare WHO life expectancy data against Numbeo healthcare index: where do they diverge?
- Rent burden analysis: rent as a share of local purchasing power
- Three side-by-side maps covering health, transport, and housing dimensions

### Section 5 : Sweet Spot Cities

- Identify cities in top quartile for quality and bottom quartile for cost
- Radar charts profiling 4 to 5 featured cities across all dimensions
- Side-by-side comparison of a best value city vs. an overhyped city

### Section 6 : Time Series

- Has cost of living risen faster than quality of life since COVID (2020)?
- Line charts tracking value index over time for selected cities
- Biggest movers: cities that have improved or declined the most

### Section 7 : Conclusion

- Answer each research question directly with one or two key findings
- Name the sweet spot cities and what they share
- Acknowledge limitations: Numbeo is crowd-sourced, WHO data is country-level not city-level, time series limited to 5 years of data
